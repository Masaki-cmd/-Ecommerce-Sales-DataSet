### KGI (Key Goal Indicator) と KPI (Key Performance Indicators) の定義

**KGI (Key Goal Indicator)** は、ビジネスの最終目標を示す指標です。通常、売上目標、利益目標、市場シェアなどが該当します。このKGIを達成するために、どのような要因が影響しているかを追跡するのがKPIです。

**KPI (Key Performance Indicators)** は、KGIの達成度を測るための具体的な中間指標です。KPIは定量的であり、進捗を測定し、戦略が成功しているかどうかを評価するために使用されます。良いKPIは、SMART（Specific, Measurable, Achievable, Relevant, Time-bound）の原則に基づいています。

ここでは、これまでのデータ分析で得られた情報と一般的なビジネス目標に基づき、KGIとKPIの候補を提案します。ユーザーはこれを基に、具体的なビジネス目標に合わせて調整してください。

### KPIツリー (KPI Tree) の構築

KPIツリーは、KGIを頂点とし、それを達成するための主要なKPIがどのように貢献し、さらにそれらのKPIがどのような下位の指標によって支えられているかを示す階層構造です。これにより、どの指標がビジネス目標に最も影響を与えるかを視覚的に理解し、戦略的な意思決定をサポートします。

前述のKGI（収益の最大化）とKPIに基づいて、以下にKPIツリーの因果構造の例を提示します。これは、指標間の関係性を示すものであり、ビジネスの特性に合わせて調整可能です。

**KGI: 収益の最大化 (Maximize Revenue)**

*   **主要な推進要因 (Primary Drivers of Revenue):**
    *   **平均販売数 (Average Units Sold)**: 直接的に収益に貢献します。
    *   **平均利益 (Average Profit)**: 収益とコスト効率の両方を示します。

*   **間接的な推進要因 (Indirect Drivers):**
    *   **平均顧客評価 (Average Customer Rating)**: 顧客満足度が高まると、リピート購入や口コミによる新規顧客獲得が増え、結果的に販売数と収益が増加する可能性があります。
    *   **広告費用対効果 (Return on Ad Spend)**: 広告活動が効率的であれば、より多くの顧客にリーチし、販売数と収益が増加する可能性があります。広告費の最適化は利益にも直結します。

## 🔷 Data Cleaning

````python
print("### 欠損処理 (Missing Value Handling) ###")

if missing_rates.sum().sum() == 0:
    print("変更内容: なし (No change) - 欠損値はありません (No missing values found).")
    print("影響件数: 0 (Affected rows: 0)")
    print("ビジネス影響: データ品質は高く、欠損データによるバイアスリスクは低い。 (Business impact: High data quality, low risk of bias due to missing data).")
else:
    # This block would contain code to handle missing values if any were present.
    print("欠損値が検出されました。分析の目的とデータの特性に基づいて処理方針を決定します。(Missing values detected. Strategy will be determined based on analysis goals and data characteristics).")
````

````python
print("### 型変換 (Type Conversion) ###")

initial_dtypes = df.dtypes.to_dict()

# Convert Month to a numerical representation and create a proper Date column
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df['Month_Num'] = df['Month'].map(month_map)
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month_Num'].astype(str) + '-01')

# Convert Category and Region to categorical type for memory efficiency and better analysis
df['Category'] = df['Category'].astype('category')
df['Region'] = df['Region'].astype('category')

# Drop original 'Month' and 'Month_Num' if 'Date' is sufficient for time-series analysis
df = df.drop(columns=['Month', 'Month_Num'])

changes_made = False
for col, new_dtype in df.dtypes.to_dict().items():
    if col in initial_dtypes and initial_dtypes[col] != new_dtype:
        print(f"  - 列 '{col}': {initial_dtypes[col]} -> {new_dtype}")
        changes_made = True
    elif col not in initial_dtypes:
        print(f"  - 新規列 '{col}': {new_dtype}")
        changes_made = True

if not changes_made:
    print("変更内容: なし (No change) - すべてのデータ型は適切です。(All data types are appropriate).")
else:
    print("変更内容: 'Month'と'Year'を結合し'Date'列を作成、'Category'と'Region'をカテゴリ型に変換。(Combined 'Month' and 'Year' to create 'Date' column, converted 'Category' and 'Region' to categorical type).")
    print(f"影響件数: {len(df)} (Affected rows: {len(df)}) - 全行のデータ型が更新されました。(Data types updated for all rows).")
    print("ビジネス影響: 時間軸での分析が可能になり、カテゴリ別・地域別の集計処理が効率化されます。(Business impact: Enables time-series analysis and efficient aggregation by category/region).")

display(df.head())
````

````python
print("### 主キーの非ユニーク性対応（集計） (Handling Non-Unique Primary Key (Aggregation)) ###")

# Re-evaluate the composite key after type conversion
composite_key_cols = ['Date', 'Category', 'Region'] # Now using the new Date column
is_unique_composite_after_conversion = df.duplicated(subset=composite_key_cols).sum() == 0

if not is_unique_composite_after_conversion:
    print("同一の年、月、カテゴリ、地域に複数のレコードが存在するため、データを集計します。\n(Multiple records exist for the same Year, Month, Category, and Region, so data will be aggregated.)")
    print(f"集計前の行数: {len(df)}")

    # Aggregate numerical columns by summing, and for Customer_Rating, we'll average it.
    # Other columns like Year, Month, Category, Region will become part of the new index or remain as grouping keys.
    aggregation_dict = {
        'Units_Sold': 'sum',
        'Revenue': 'sum',
        'Profit': 'sum',
        'Customer_Rating': 'mean', # Assuming rating should be averaged if multiple exist for a segment
        'Ad_Spend': 'sum'
    }

    # Fix: Use reset_index() explicitly and add observed=True for categorical performance
    df_agg = df.groupby(composite_key_cols, observed=True).agg(aggregation_dict).reset_index()

    # Update df to the aggregated DataFrame
    affected_rows_count = len(df) - len(df_agg)
    df = df_agg

    print(f"変更内容: 'Date', 'Category', 'Region'で重複するレコードを集計。(Aggregated duplicate records by 'Date', 'Category', 'Region').")
    print(f"影響件数: {affected_rows_count}行が結合されました。(Affected rows: {affected_rows_count} rows were combined).")
    print(f"集計後の行数: {len(df)}")
    print("ビジネス影響: データが正しい粒度で構造化され、より正確な分析と意思決定が可能になります。(Business impact: Data is structured at the correct granularity, enabling more accurate analysis and decision-making).")
else:
    print("変更内容: なし (No change) - 'Date', 'Category', 'Region'の組み合わせはユニークです。(The combination of 'Date', 'Category', 'Region' is unique).")
    print("影響件数: 0 (Affected rows: 0)")
    print("ビジネス影響: データの粒度はすでに分析に適しています。(Business impact: Data granularity is already suitable for analysis).")

display(df.head())
````

````python
print("### 外れ値検証（IQR） (Outlier Verification (IQR)) ###")

numerical_cols = ['Units_Sold', 'Revenue', 'Profit', 'Customer_Rating', 'Ad_Spend']
outlier_info = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    if not outliers.empty:
        outlier_info[col] = {
            'count': len(outliers),
            'percentage': len(outliers) / len(df) * 100,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound
        }

if outlier_info:
    print("外れ値が検出されました。(Outliers detected).")
    for col, info in outlier_info.items():
        print(f"  - 列 '{col}': {info['count']} 件 ({info['percentage']:.2f}%) の外れ値 (Outliers: {info['count']} ({info['percentage']:.2f}%))")
        print(f"    - 許容範囲: [{info['lower_bound']:.2f}, {info['upper_bound']:.2f}] (Acceptable range: [{info['lower_bound']:.2f}, {info['upper_bound']:.2f}])")
    print("変更内容: なし (No change) - 外れ値は報告のみ。(Outliers are only reported).")
    print(f"影響件数: {sum(info['count'] for info in outlier_info.values())} (Affected rows: {sum(info['count'] for info in outlier_info.values())})")
    print("ビジネス影響: これらの外れ値は、特別なプロモーション、季節的なピーク、または記録エラーを示している可能性があります。さらなる調査が必要になる場合があります。(Business impact: These outliers may indicate special promotions, seasonal peaks, or recording errors. Further investigation may be required).")
else:
    print("変更内容: なし (No change) - 外れ値は検出されませんでした。(No outliers detected).")
    print("影響件数: 0 (Affected rows: 0)")
    print("ビジネス影響: データの分布は比較的均一で、極端な異常値のリスクは低い。(Business impact: Data distribution is relatively uniform, with low risk of extreme outliers).")
````

````python
print("### 異常値チェック（業務ロジック） (Anomaly Check (Business Logic)) ###")

anomalies_found = False
anomaly_count = 0

# Check for non-negative values where applicable
non_negative_cols = ['Units_Sold', 'Revenue', 'Profit', 'Ad_Spend']
for col in non_negative_cols:
    negative_values = df[df[col] < 0]
    if not negative_values.empty:
        print(f"  - 列 '{col}': {len(negative_values)} 件の負の値が検出されました。(Negative values detected: {len(negative_values)} in column '{col}').")
        anomaly_count += len(negative_values)
        anomalies_found = True

# Check Customer_Rating range (assuming 1-5 scale)
if 'Customer_Rating' in df.columns:
    invalid_ratings = df[(df['Customer_Rating'] < 1) | (df['Customer_Rating'] > 5)]
    if not invalid_ratings.empty:
        print(f"  - 列 'Customer_Rating': {len(invalid_ratings)} 件の範囲外の値が検出されました。(Out-of-range values detected: {len(invalid_ratings)} in column 'Customer_Rating').")
        anomaly_count += len(invalid_ratings)
        anomalies_found = True

if anomalies_found:
    print("変更内容: なし (No change) - 異常値は報告のみ。(Anomalies are only reported).")
    print(f"影響件数: {anomaly_count} (Affected rows: {anomaly_count})")
    print("ビジネス影響: これらの異常値はデータ入力エラー、またはビジネスプロセスの予期せぬ挙動を示している可能性があります。原因調査が必要です。(Business impact: These anomalies may indicate data entry errors or unexpected business process behavior. Root cause investigation is needed).")
else:
    print("変更内容: なし (No change) - 業務ロジックに基づく異常値は検出されませんでした。(No anomalies detected based on business logic).")
    print("影響件数: 0 (Affected rows: 0)")
    print("ビジネス影響: データは業務ロジックに沿っており、信頼性は高い。(Business impact: Data aligns with business logic and is highly reliable).")
````

````python
print("### データ品質スコア (Data Quality Score) ###")

# Calculate Data Quality Score (0-100)
# Start with a base score and deduct points for issues found.
quality_score = 100
reasons = []

if missing_rates.sum().sum() > 0:
    quality_score -= 10 # Deduct for missing values
    reasons.append("欠損値の存在 (Presence of missing values)")

if duplicate_rows > 0:
    quality_score -= 5 # Deduct for duplicate rows
    reasons.append("重複行の存在 (Presence of duplicate rows)")

# Check for uniqueness after aggregation for the composite key
composite_key_cols_final = ['Date', 'Category', 'Region']
is_unique_composite_final = df.duplicated(subset=composite_key_cols_final).sum() == 0
if not is_unique_composite_final: # This should ideally be True after aggregation
    quality_score -= 15 # Significant deduction if still not unique after aggregation attempt
    reasons.append("集計後も主キー候補がユニークではない (Composite primary key candidate still not unique after aggregation)")

if 'outlier_info' in locals() and outlier_info: # Check if outlier_info was populated
    total_outliers_percentage = sum(info['percentage'] for info in outlier_info.values())
    quality_score -= min(10, int(total_outliers_percentage * 0.5)) # Deduct based on percentage of outliers, max 10 points
    reasons.append("外れ値の存在 (Presence of outliers)")

if 'anomalies_found' in locals() and anomalies_found:
    quality_score -= 15 # Deduct for business logic anomalies
    reasons.append("業務ロジックに基づく異常値の存在 (Presence of anomalies based on business logic)")

# Ensure score doesn't go below 0
quality_score = max(0, quality_score)

print(f"データ品質スコア: {quality_score}/100 (Data Quality Score: {quality_score}/100)")
if reasons:
    print("主な理由: (Main reasons:)")
    for r in reasons:
        print(f"  - {r}")
else:
    print("特筆すべきデータ品質上の問題はありません。(No significant data quality issues).")

print("\nデータクリーニングフェーズが完了しました。(Data Cleaning phase completed.)")
````

### カテゴリ別の収益性と広告費用対効果（ROAS）の相関

KGI「収益の最大化」において、利益率と広告効率は非常に重要な指標です。ここでは、各カテゴリの平均利益とROAS（Return on Ad Spend）の関係を可視化し、どのカテゴリが効率的に収益を上げているかを分析します。ROASは`Revenue / Ad_Spend`として計算します。

````python
# Calculate ROAS for each category
# Avoid division by zero: if Ad_Spend is 0, ROAS is undefined or very high. Handle this by replacing 0 with a small number or setting ROAS to NaN/inf.
# For simplicity, if Ad_Spend is 0, we'll set ROAS to NaN or treat it as a very high number that should be investigated.
category_summary['ROAS'] = category_summary['Revenue'] / category_summary['Ad_Spend']

plt.figure(figsize=(12, 8))
sns.scatterplot(x='ROAS', y='Profit', hue='Category', data=category_summary, s=200, alpha=0.7)

# Add labels for each point
for i, row in category_summary.iterrows():
    plt.text(row['ROAS'] + 0.5, row['Profit'], row.name, horizontalalignment='left', verticalalignment='center')

plt.title('Category Performance: Average Profit vs. ROAS')
plt.xlabel('Return on Ad Spend (ROAS)')
plt.ylabel('Average Profit')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("ビジネス影響: この散布図は、各カテゴリの平均利益とROASの関係を明確に示します。右上にあるカテゴリは、高い利益率と広告効率の両方を達成しており、ビジネスにとって理想的なカテゴリです。左下にあるカテゴリは、改善の機会が大きいことを示唆します。これにより、広告予算の配分やカテゴリ別の戦略立案に役立つ洞察が得られます。(Business impact: This scatter plot clearly shows the relationship between average profit and ROAS for each category. Categories in the upper right quadrant achieve both high profitability and advertising efficiency, making them ideal for the business. Categories in the lower left suggest significant room for improvement. This provides insights useful for allocating advertising budgets and formulating category-specific strategies.)")
````

### Booksカテゴリ ROAS算出と影響要因の統計分析 (ROAS Calculation and Statistical Analysis of Influencing Factors for Books Category)

BooksカテゴリのROAS改善に向けた具体的な広告施策を提案するにあたり、まずBooksカテゴリにおけるROASの算出と、それに影響を与える要因の統計分析を行います。

#### 1. Booksカテゴリ内でのROAS算出 (ROAS Calculation within Books Category)

各データポイント（年、月、地域）ごとのROASを`Revenue / Ad_Spend`として算出します。`Ad_Spend`が0の場合は分析に影響を与えないよう適切に処理します。

````python
# Calculate ROAS for each entry in the df_books DataFrame
# Handle cases where Ad_Spend might be zero to avoid division by zero errors.
# If Ad_Spend is 0, ROAS is effectively infinite, but for practical analysis, we can assign a large finite number or NaN.
# Based on df_books.describe(), min Ad_Spend is not zero, so direct division should be fine.

df_books_analyzed = df_books.copy()
df_books_analyzed['ROAS'] = df_books_analyzed['Revenue'] / df_books_analyzed['Ad_Spend']

# Display summary statistics for the newly calculated ROAS
print("### Booksカテゴリ内のROAS統計量 (ROAS Statistics within Books Category) ###")
display(df_books_analyzed['ROAS'].describe())

print("ビジネス影響: Booksカテゴリ内のROASの平均、中央値、分布を理解することで、広告投資が収益にどの程度効率的に結びついているかについての基礎的な理解が得られます。これにより、カテゴリ全体の広告効率の健全性を評価できます。(Business impact: Understanding the mean, median, and distribution of ROAS within the Books category provides a baseline understanding of how efficiently ad spend translates into revenue, allowing for an assessment of the overall advertising efficiency of the category.)")
````

#### 2. ROASと他の数値変数との相関分析 (Correlation Analysis between ROAS and other Numerical Variables)

ROASに影響を与える可能性のある要因（Units_Sold, Profit, Customer_Rating）との相関を分析することで、ROAS改善のための具体的なアプローチを特定します。

````python
numerical_cols_for_roas_corr = ['Units_Sold', 'Profit', 'Customer_Rating', 'ROAS']
correlation_matrix_roas = df_books_analyzed[numerical_cols_for_roas_corr].corr()

print("### Booksカテゴリ内のROASと主要数値変数の相関行列 (Correlation Matrix of ROAS and Key Numerical Variables in Books Category) ###")
display(correlation_matrix_roas)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_roas, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Books Category: Correlation Matrix with ROAS')
plt.tight_layout()
plt.show()

print("ビジネス影響: ROASと他の主要指標（Units_Sold, Profit, Customer_Rating）との相関を分析することで、どのビジネス活動がROASに最も強く結びついているかを特定できます。高い正の相関は、その指標の改善がROAS向上に直結する可能性が高いことを示唆します。(Business impact: Analyzing the correlation between ROAS and other key metrics helps identify which business activities are most strongly linked to ROAS. A strong positive correlation suggests that improving that metric is highly likely to directly improve ROAS.)")
````

#### 3. Booksカテゴリにおける地域別ROAS分析 (Regional ROAS Analysis within Books Category)

````python
books_roas_by_region = df_books_analyzed.groupby('Region')['ROAS'].mean().sort_values(ascending=False)

print("### Booksカテゴリ 地域別平均ROAS (Average ROAS by Region for Books Category) ###")
display(books_roas_by_region.to_frame())

plt.figure(figsize=(10, 6))
sns.barplot(x=books_roas_by_region.index, y=books_roas_by_region.values, palette='viridis')
plt.title('Books Category: Average ROAS by Region')
plt.xlabel('Region')
plt.ylabel('Average ROAS')
plt.tight_layout()
plt.show()

print("ビジネス影響: Booksカテゴリ内での地域別ROASの比較は、広告戦略の効率が地域によって大きく異なることを示唆します。ROASが低い地域は、広告ターゲットの再評価、チャネルの見直し、クリエイティブのローカライズなど、地域に特化した広告最適化の緊急性が高いことを意味します。(Business impact: Comparing regional ROAS within the Books category suggests significant variations in advertising strategy effectiveness across regions. Regions with lower ROAS indicate an urgent need for region-specific advertising optimization, such as re-evaluating ad targeting, reviewing channels, or localizing creatives.)")
````

*   **因果関係の特定**: 相関分析は因果関係を示すものではないため、施策の導入に際してはA/Bテストなどによる効果検証が不可欠です。
*   **外れ値の影響**: 外れ値は検出されましたが除去されていないため、一部の統計的要約や平均値に影響を与えている可能性があります。
*   **データ期間**: 提供されたデータは特定の期間のものであるため、長期的なトレンドや季節性の変化をより正確に捉えるには、より長期間のデータが必要となる場合があります。
